```python
import numpy as np
from tabulate import tabulate;

def fact(n):
    f = 1.
    for i in range (1,n+1):
        f *= i
    return (f)

def binom(n,k):
    f = fact(n)/(fact(n-k)*fact(k))
    return (f)

def bernoulli_calc(n):
    if (n<=61):
        bn = 0.
        for k in range(n+1):
            s = 0.
            for j in range(k+1):
                s += ((-1)**j) * ((j)**n) * binom(k,j)
            bn += s/(k+1)
        return (bn)
    else:
        return (np.nan)
    
def bernoulli_lookup(n):
    
# drawn from https://codegolf.stackexchange.com/questions/65382/bernoulli-numbers
# the sign of bn[1] is reversed for consistency with current convention
    
    bn=np.zeros(61)

    bn[0] = 1.
    bn[1] = -0.5
    bn[2] = 0.166666666666667
    bn[4] = -0.0333333333333333
    bn[6] = 0.0238095238095238
    bn[8] = -0.0333333333333333
    bn[10] = 0.0757575757575758
    bn[12] = -0.253113553113553
    bn[14] = 1.16666666666667
    bn[16] = -7.0921568627451
    bn[18] = 54.9711779448622
    bn[20] = -529.124242424242
    bn[22] = 6192.1231884058
    bn[24] = -86580.2531135531
    bn[26] = 1425517.16666667
    bn[28] = -27298231.0678161
    bn[30] = 601580873.900642
    bn[32] = -15116315767.0922
    bn[34] = 429614643061.167
    bn[36] = -13711655205088.3
    bn[38] = 488332318973592
    bn[40] = -19296579341940100
    bn[42] = 841693047573682000
    bn[44] = -40338071854059400000
    bn[46] = 2.1150748638082E+21
    bn[48] = -1.20866265222965E+23
    bn[50] = 7.50086674607695E+24
    bn[52] = -5.03877810148106E+26
    bn[54] = 3.6528776484818E+28
    bn[56] = -2.84987693024508E+30
    bn[58] = 2.38654274996836E+32
    bn[60] = -2.13999492572252E+34

    if (n<=61):
        bernn = bn[n]
        
    else:
        bernn = np.nan
    return (bernn)  
    
def taylorTan(n,x):
    sum = 0
    for i in range (1,n+1):
        sum += bernoulli_lookup(2*i) * ((-4)**i) * (1-(4**i)) * (x**(2*i-1)) / fact(2*i)
    return sum

def min_n(a):
    for i in range (0,31):
        if a[i] < 10**-12:
            return i
    return 31

def main():
    arr = [0*math.pi/16,1*math.pi/16,2*math.pi/16,3*math.pi/16,4*math.pi/16,5*math.pi/16,6*math.pi/16,7*math.pi/16,8*math.pi/16,9*math.pi/16,10*math.pi/16,11*math.pi/16,12*math.pi/16,13*math.pi/16,14*math.pi/16,15*math.pi/16]
    diff_k0 = []
    diff_k1 = []
    diff_k2 = []
    diff_k3 = []
    diff_k4 = []
    diff_k5 = []
    diff_k6 = []
    diff_k7 = []
    diff_k8 = []
    diff_k9 = []
    diff_k10 = []
    diff_k11 = []
    diff_k12 = []
    diff_k13 = []
    diff_k14 = []
    diff_k15 = []
    min_n_ind = []
    min_val = []
    tan_exact = []
    diff = []
        
    for n in range (0,31):
        diff_k0.append(abs(taylorTan(n,arr[0]) - math.tan(arr[0])))
        diff_k1.append(abs(taylorTan(n,arr[1]) - math.tan(arr[1])))
        diff_k2.append(abs(taylorTan(n,arr[2]) - math.tan(arr[2])))
        diff_k3.append(abs(taylorTan(n,arr[3]) - math.tan(arr[3])))
        diff_k4.append(abs(taylorTan(n,arr[4]) - math.tan(arr[4])))
        diff_k5.append(abs(taylorTan(n,arr[5]) - math.tan(arr[5])))
        diff_k6.append(abs(taylorTan(n,arr[6]) - math.tan(arr[6])))
        diff_k7.append(abs(taylorTan(n,arr[7]) - math.tan(arr[7])))
        diff_k8.append(abs(taylorTan(n,arr[8]) - math.tan(arr[8])))
        diff_k9.append(abs(taylorTan(n,arr[9]) - math.tan(arr[9])))
        diff_k10.append(abs(taylorTan(n,arr[10]) - math.tan(arr[10])))
        diff_k11.append(abs(taylorTan(n,arr[11]) - math.tan(arr[11])))
        diff_k12.append(abs(taylorTan(n,arr[12]) - math.tan(arr[12])))
        diff_k13.append(abs(taylorTan(n,arr[13]) - math.tan(arr[13])))
        diff_k14.append(abs(taylorTan(n,arr[14]) - math.tan(arr[14])))
        diff_k15.append(abs(taylorTan(n,arr[15]) - math.tan(arr[15])))

    min_n_ind.append(min_n(diff_k0))
    min_n_ind.append(min_n(diff_k1))
    min_n_ind.append(min_n(diff_k2))
    min_n_ind.append(min_n(diff_k3))
    min_n_ind.append(min_n(diff_k4))
    min_n_ind.append(min_n(diff_k5))
    min_n_ind.append(min_n(diff_k6))
    min_n_ind.append(min_n(diff_k7))
    min_n_ind.append(min_n(diff_k8))
    min_n_ind.append(min_n(diff_k9))
    min_n_ind.append(min_n(diff_k10))
    min_n_ind.append(min_n(diff_k11))
    min_n_ind.append(min_n(diff_k12))
    min_n_ind.append(min_n(diff_k13))
    min_n_ind.append(min_n(diff_k14))
    min_n_ind.append(min_n(diff_k15))
    
    for n in range (0,16):
        min_val.append(taylorTan(min_n_ind[n],arr[n]))
        
    for n in range (0,16):
        tan_exact.append(math.tan(arr[n]))
    
    for n in range (0,16):
        diff.append(abs(min_val[n] - tan_exact[n]))
        
    table=[["Exact Value",tan_exact[0],tan_exact[1],tan_exact[2],tan_exact[3],tan_exact[4],tan_exact[5],tan_exact[6],tan_exact[7],tan_exact[8],tan_exact[9],tan_exact[10],tan_exact[11],tan_exact[12],tan_exact[13],tan_exact[14],tan_exact[15]],
           ["Taylor Sum",min_val[0],min_val[1],min_val[2],min_val[3],min_val[4],min_val[5],min_val[6],min_val[7],min_val[8],min_val[9],min_val[10],min_val[11],min_val[12],min_val[13],min_val[14],min_val[15]],
           ["Error",diff[0],diff[1],diff[2],diff[3],diff[4],diff[5],diff[6],diff[7],diff[8],diff[9],diff[10],diff[11],diff[12],diff[13],diff[14],diff[15]],
           ["Min n Value",min_n_ind[0],min_n_ind[1],min_n_ind[2],min_n_ind[3],min_n_ind[4],min_n_ind[5],min_n_ind[6],min_n_ind[7],min_n_ind[8],min_n_ind[9],min_n_ind[10],min_n_ind[11],min_n_ind[12],min_n_ind[13],min_n_ind[14],min_n_ind[15]]]
    print(tabulate(table,headers=["Labels","K=0","K=1","K=2","K=3","K=4","K=5","K=6","K=7","K=8","K=9","K=10","K=11","K=12","K=13","K=14","K=15"],disable_numparse=True))
main()
```
